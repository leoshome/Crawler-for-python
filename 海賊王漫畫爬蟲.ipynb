```python
def getPicture():
#    urlRow = http://www.dbmeizi.com/category/2?p=4#
    pictureNum = 0
    for num in range(2, 40):
        meyUrl = 'http://www.dbmeizi.com/category/2?p=' + str(num) + '#'
        url = requests.get(meyUrl)
        soup = BeautifulSoup(url.text)
        firstLst = soup.findAll('li', attrs={'class':'span3'})
        for line in firstLst:
            for yes in line.findAll('img'):
                print(yes.get('data-src') + '    id號碼: ' + yes.get('data-id') +  yes.get('data-title'))
                urlretrieve(yes.get('data-src'), '/Users/OSX/Desktop/pictures/'+yes.get('data-id')+'.jpg')
                pictureNum += 1
    print(pictureNum)
getPicture()
```

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.jkforum.net/forum-574-1.html'
soup = BeautifulSoup(requests.get(url).text)

def downloadPic(urls, num):
    soup = BeautifulSoup(requests.get(urls).text)
    for line in soup.select('.zoom'):
        num += 1
        try:
            urlretrieve(line.get('zoomfile') , '/Users/OSX/Desktop/pic/' + str(num) +'.jpg')
            print(line.get('zoomfile'))
            print(num)
        except:
            num += 1
            pass

global num 
num = 0
for line in soup.select('.new'):
    num += 1
    try:
        text = line.select('.xst')[0]
        href = line.select('a')[0]
        print(text.string, 'http://www.jkforum.net/'+href['href'])
        downloadPic('http://www.jkforum.net/'+href['href'], num)
    except:
        pass
```

#下載海賊王

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.cartoonmad.com/comic/1152.html'
commicPage = {}

#處理集數的字串
def stripText(line):
    if "卷" in line.text:
        a, b, c = line.text.split()
        return(a+b+c)
    else:
        return ('話數')
    
#卷數處理
def openCommic (openUrl):
    soup = BeautifulSoup(urlopen(url).read())
    commicBook = {}
    for line in soup.findAll('a', attrs={'target': '_blank'}):
        try:
            if 'comic' in 'http://web.cartoonad.com/' + line['href']:
                commicBook[stripText(line)] = 'http://web.cartoonad.com/' + line['href']
        except:
            pass
    return commicBook

#頁次處理
def pageComic(commicDic):
    commicLst =sorted(commicDic.items())
    for line in commicLst:
        soup = BeautifulSoup(urlopen(line[1]).read())
        num = 1
        for lineNext in soup.select('option'):
            if lineNext.get('value') != None:
                strs =  (lineNext.get('value')) 
                pageUrl = 'http://web.cartoonad.com//comic/' + strs
                commicNum = line[0] + ':第' + str(num) + '頁'
                downloadCommic(pageUrl, commicNum)
                num += 1
                
#下載漫畫                
def downloadCommic(pageUrl, commicNum):
    soup = BeautifulSoup(urlopen(pageUrl).read())
    for line in soup.select('img'):
        if 'jpg' in line['src']:
            print(commicNum + line['src'])
            urlretrieve(line['src'], '/Users/OSX/Desktop/pic/' + commicNum + '.jpg' )

dic = openCommic(url)
pageComic(dic)
```
